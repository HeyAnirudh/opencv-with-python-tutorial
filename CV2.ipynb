```python
import numpy as np

import cv2
from google.colab.patches import cv2_imshow
```

```python
imag=cv2.imread("/demo-image2.png")
gray=cv2.cvtColor(imag,cv2.COLOR_BGR2GRAY)
can=cv2.Canny(gray,150,120)
kernal=np.ones((5,5),np.uint8)
erode=cv2.erode(imag,kernal,iterations =1)
dil=cv2.dilate(imag,kernal,iterations=1)
HOL=np.hstack((dil,erode))
LOL=np.hstack((can,gray))
cv2_imshow(HOL)
cv2_imshow(LOL)
print(erode.shape)
print(can.shape)
print(dil.shape)
```

```python
import cv2 
from google.colab.patches import cv2_imshow
import numpy as np
img=cv2.imread("/shapes.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
ret,thres=cv2.threshold(gray,50,255,1)
cv2_imshow(thres)
```

```python
counters,h=cv2.findContours(thres,1,2)
print(counters)
```

```python
for count in counters:
  approx=cv2.approxPolyDP(count,0.01*cv2.arcLength(count,True),True)
  n=len(approx)
  if n==4:
     print("we have a square")
     cv2.drawContours(img,[count],0,255,10)
 
  elif n==6:
     print("we have a hexagon")
     cv2.drawContours(img,[count],0,(0,255,124),4)
  
cv2_imshow(img)
```
